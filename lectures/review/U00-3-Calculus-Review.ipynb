# Functions and Graphs

We have been using functions in both the mathematical sense and the programming sense extensively in this
class.  In the mathematical sense, the simplest function is a function of a single independent variable
$x$.  We often also use a dependent variable $y$ and say that "y is a **function** of x", written this
way

\begin{equation}
y = f(x)
\end{equation}

The simplest function we can imagine is a constant function, for example, 

\begin{equation}
y = 3
\end{equation}

Here you may ask how is this a function of the independent variable $x$?  In this case it is not, $y$ does
not depend on $x$, although we can also think of it like this:

\begin{equation}
y = 0 \cdot x + 3
\end{equation}

We could also graph this constant function.  It is a pretty uninteresting plot, since the value of $y$ is 
always 3 no matter what the value of $x$ is:

```python
# Example of constant function, it doesn't matter what the
# independent variable x value is, y is always a constant 3
# y = 3
def f(x):
    return 0*x + 3


x = np.linspace(-3, 3)
plt.plot(x, f(x));
plt.ylim([-1, 4])
plt.grid();
```

## Slope of a Line and the Slope-Intercept Equation

The slope of a line is simply the ratio of the change in $y$, written as $\Delta y$ to the change in $x$, $\Delta x$

\begin{equation}
m = \frac{\Delta y}{\Delta x}
\end{equation}

(m is often used to represent the slop of a line).  For our previous equation $y = 0x + 3$ you
should see that the slope of the line is 0, because $y$ never changes
thus the ratio is 0.  We say that the line has a slope of 0 in this case. 

Lets try a different function, say

\begin{equation}
y = 2x
\end{equation}

Here is the graph of this line.

```python
def f(x):
    return 2*x
    
x = np.linspace(-3, 3)
plt.plot(x, f(x));
plt.xlim([-6, 6])
plt.grid();
```

Here the slope $m = 2$ because for every change of $x$ $\Delta x = 1$, $y$ will change by 2 $\Delta y = 2$
and thus the ratio is $m = \Delta y / \Delta x = 2 / 1 = 2$

If we add another term to the equation

\begin{equation}
y = 2x + 3
\end{equation}

The graph of the equation now looks like this:

```python
def f(x):
    return 2*x + 3

x = np.linspace(-3, 3)
plt.plot(x, f(x));
plt.xlim([-6, 6])
plt.grid();
```

Here the slope of the line is still the same $m = 2$.  But when $x = 0$ $y$ has a value of $3$.
Whenever you rearrange a linear function into the form $y = mx + b$, this is known as the slope-intercept
form of the equation of the line.  $m$ will be the slope, which gives the ratio of how $y$ changes to the amount
that $x$ changes.  The constant $b$ will be the intercept of the line.  When $x = 0$ this will be the value of $y$
or in other words this will be the intercept location, the value of $y$ when x changes from negative to positive (at
the point $x = 0$ ).

At this point you should understand the following about the slope and the equation of a line

- A line always has a constant slope.  At all places on the line, the slope is always the same $m$.  The
  slope never changes.
- The slope-intercept equation of a line allows you to directly see the slope of the line and the intercept
  location where the line crosses the axis at $x = 0$.
- A positive slope, like $+2$ above will describe a line that goes up, as $x$ gets bigger $y$ gets bigger.  You can of
  course have a negative slope, in which case the line goes down, as $x$ gets bigger $y$ decreases.
- A slope of 0 is a constant function, and defines a flat line as seen on its graph.

no longer have a constant slope.  The rate of change of the line at any point $x$ of the function will
depend on the function properties at that point.  

Lets say we want to know what the slope is of the basic quadratic function $y = x^2$ is at the point where $x = 2$.
We could use the equation for the slope of a line, and calculate the ratio of the change of $y$ to $x$

You should intuitively see that as we make the change in $x$ smaller ($\Delta x$) the lines they define that we
measure the slope of are approaching the red line.  The red line represents the tangent to the curve right at the point
we want to measure the slope $x = 2$.  For a quadratic curve like this one, the slope is not constant, it is constantly
changing.  Each location on the equation will have a slightly different slope.  But we can measure the
"instantaneous slope", also called the "instantaneous rate of change" at a given point.  This is simply the slope
of the red line shown.  More formally, in the limit as $\Delta x$ gets smaller and smaller and eventually reaches
0, the lines we measure the slope of get closer and closer to the red line, the tangent line which is the true
instantaneous slope of the equation at the point $x = 2$.  Formally we can use the concept of a limit to specify
this idea:

\begin{equation}
m_{\text{tan}} = \lim\limits_{\Delta x \to 0} \frac{\Delta y}{\Delta x} = \lim\limits_{\Delta x \to 0} \frac{f(x + \Delta x) - f(x)}{\Delta x} 
\end{equation}

This idea, denoted by the notation $\frac{dy}{dx}$ is called the **derivative** of the function 
$y = f(x)$ measured instantaneously at a given point $x$.  The derivative is another name for this notion
of the measure of the instantaneous rate of change, or the slope of the function at a particular point on
the function.

There are many different notations to represent this concept of the derivative:
$\frac{d f(x)}{dx}$, $\frac{d}{dx} f(x)$, $f'(x)$. (The last one is sometimes referred to as "prime notation.") 

It is possible to come up with a formula that can be used to calculate the derivitaive of a function at
any point.  For example, lets look quickly at the function $y = f(x) = x^2$ above. For any value of $x$,
using the definition above, we have the following expression

\begin{equation}
m = \frac{\Delta y}{\Delta x} 
  =  \frac{(x + \Delta x)^2 - x^2}{\Delta x} 
  = \frac{x^2 + 2x \Delta x + (\Delta x)^2 - x^2}{\Delta x}
  = \frac{\Delta x (2x + \Delta x)}{\Delta x}
  = 2x + \Delta x
\end{equation}

The previous sequence may look complicated, but we are using simple algebra to expand and cancel out terms, just
with the maybe unfamiliar term $\Delta x$ in the expressions.  The result of this rearrangement though shows
something interesting.  This says simply that the slope of any line at any point of the function
is $2x + \Delta x$ for a given $\Delta$.  But also it says, in the limit, when $\Delta x = 0$, the
instantaneous slope will be simply $2x$.  Or in other words we have shown that for our function $f(x) = x^2$

\begin{equation}
\frac{d f(x)}{dx} = 2x
\end{equation}

This means that the derivative (instantaneous rate of change) of the function is $2x$ at any point $x$
of this function.  We can use this expression for the derivative of our function $y = x^2$
to determine the instantaneous rate of change at any point of our function.  For example:

```python
def f(x):
    """The simple quadratic function we are using y = x^2
    """
    return x**2

def df(x):
    """The derivative of our equation, can be used to determine the instantaneous
    rate of change, or in other words the slope, of the equation at any particular
    point x.
    """
    return 2 * x;

def plot_tangent_line(x):
    """A little utility function, this will plot a tangent line
    on the current figure at the location x, making use of the
    slope of the function f(x) using its derivative df(x), and
    plotting a line that goes through x with that slope
    """
    # determine value of y at the point x, and the slope of the line
    # at point x
    y = f(x)
    m = df(x)
    
    # y = mx + b
    # we have y and m for our needed line, need to determine
    # correct value of be, so rearrange
    # b = y - mx
    # and determine the correct intercept of the tangent line
    b = y - m * x
    
    # we will plot a line from -0.5 to +0.5 of x
    deltax = 0.75
    x0 = x - deltax
    x1 = x + deltax
    
    # we have the equation of the line, and we have two point x0, x1, determine
    # the points y0, y1 on the line
    y0 = m * x0 + b
    y1 = m * x1 + b
    
    # now we can plot the tangent line
    plt.plot([x0, x1], [y0, y1], 'r--')
    
    # plot a marker at the tangent point
    plt.plot(x, y, 'ro')
    
    # annotate the plot
    plt.text(x+0.05, y, 'x = %f\nm = %f' % (x, m))
    
    
# plot the original function
x = np.linspace(-5.0, 5.0)
plt.plot(x, f(x))
plt.grid();

# plot some tangent lines at different points, using the derivative (slope) of
# the function to do it.  We use a small utility function to plot several
# example tangent lines
plot_tangent_line(-4)
plot_tangent_line(-2)
plot_tangent_line(0)
plot_tangent_line(2)
plot_tangent_line(4)
```

# General Rules of Derivation

Using similar methods to this, we can also analyze functions like $y = x^3$ or $y = x^4$.  In general if you do
this you can easily see that a general rule exists for any power function $x^p$.  Namely if $y = f(x) = x^p$
then

**Power Rule**

\begin{equation}
\frac{d}{dx} (x^p) = p x^{p-1}
\end{equation}

This means that for example for the function $y = x^5$ the derivative of this function can be calculated
as

\begin{equation}
\frac{dy}{dx} = f'(x) = 5 x^{4}
\end{equation}

Likewise through similar analysis it is possible to derive other general rules of derivation.  Here are a few
that are useful:

**Exponential Rule**

If $y = f(x) = b^x$ then

\begin{equation}
\frac{d}{dx} (b^x) = b^x \ln(b)
\end{equation}

If $y = f(x) = e^{ax}$ then

\begin{equation}
\frac{d}{dx} (e^{ax}) = a e^{ax}
\end{equation}


**Logarathim Rule**

if $y = f(x) = \log_b(x)$ then

\begin{equation}
\frac{d}{dx} (\log_b(x)) = \frac{1}{x} \frac{1}{\ln(b)}
\end{equation}

as a special case, when $(b = e)$ then

\begin{equation}
\frac{d}{dx} (\ln(x)) = \frac{1}{x}
\end{equation}

Given some general rules like this, we can define several properties of derivatives, that allow us to

# Applications of Derivatives: Roots and Minima / Maxima

## Finding roots of an Equation

It is often useful to find what are known as the **roots** of an equation, i.e. the location where a function
$f(x) = 0$, or in other words the values where the graph of $f(x)$ intersects the x-axis.  Algebraically this can
be extremely tedious or even impossible, so we often use numerical techniques.  We will later look at a few such
root finding techniques.  The most basic technique known as the Newton method, or the Netwon-Raphson method, starts
with an initial guess $x_0$ and then produces a sequences of values (i.e.
it defines a series) that converges to a numerical solution
of a root.  By iterating the expression

\begin{equation}
x_i = x_{i-1} - \frac{f(x_{i-1})}{f'(x_{i-1})}
\end{equation}

Take for example the function

\begin{equation}
f(x) = x^3 - 21 x^2 + 135x - 220
\end{equation}


If we explore the graph of this funtion, it appears to have a root somewhere between 2.4 and 2.6

```python
def f(x):
    return x**3 - 21 * x**2 + 135 * x - 220

x = np.linspace(2.0, 3.0)
plt.plot(x, f(x))
plt.grid();
```

Starting with an initial value of $x_0 = 2$ we can iterate using the previous described Newton's method
like this:

```python
def f(x):
    return x**3 - 21 * x**2 + 135 * x - 220

def df(x):
    """Derivative of f(x)
    """
    return 3 * x**2 - 42 * x + 135

NUM_ITER = 10
x = np.zeros(NUM_ITER)
x[0] = 2.0
print(0, x[0])
for i in range(1, NUM_ITER):
    x[i] = x[i-1] - f(x[i-1]) / df(x[i-1])
    print(i, x[i])
```

We can confirm that this final value in the sequence is indeed a root of the equation
by plugging it back into the function.

```python
print(x[9], f(x[9]))
```

Since this is a numerical method, the result won't be exactly 0, but the function
should essentially be very close to 0 at the point we numerically determined is a
root.  You can also check the graph of the function below and see that the function
crosses the x axis at this value of x.

To understand why this works, look at the expression for the series.  We are
subtracting the expression

\begin{equation}
\frac{f(x)}{f'(x)}
\end{equation}

each time to come up with the next value in the series.  Remember we are trying to find
the root of the equation $y = f(x)$.  So when $f(x)$ is 0 or close to 0, this
expression is going to be 0 or close to 0, and we won't be subtracting much from the
current value in the series.  

Also we need to combine information about whether $f(x)$ is positive or negative
along with information about whether the derivative (slope) is positive or negative
to determine whether we want to make $x$ bigger or smaller to move in a
direction closer to a root.  Look at the figure for our example function.  At the
starting point of $x_0 = 2$, the function $f(x) < 0$  thus $f(x)$ is negative.
The root in our vicinity is to the right (we need $x$ to increase).  The fact that
the derivative (slope) is positive at the point $x_0 = 2$ means that if we increase 
$x$ we should be moving up the curve to a location closer to where $f(x) = 0$.  Thus
by combining information about the sign of $f(x)$ and the sign of the slope
$f'(x)$ this lets us determine if we need to increase or decrease our next value
of $x$.  In this particular case, since $f(x)$ is negative and the derivative $f'(x)$
is positive, the expression ends up being negative, and since we are subtracting a negative
we end up increasing $x$ in the next step.  If instead we had started at
$x_0 = 3$, the function $f(x)$ would be positive at this location, and the derivative
would also be positive, which would indicate we need to subtract some amount from
$x$ to go towards the root.  The same argument/intuition can be made for when the
derivative is negative, in which case which direction you need to go to approach the
root will be reversed from our example here, but again the general Newton's method
expression will work to move you closer to the root of the equation.


## Minimum and Maximum of Equations

Often it is useful to be able to "minimize" or "maximize" a function.  This means
that we want to find the location where the function is at its largest or smallest
value (often restricted to some range).

A function will be at a local maximum or minimum when its derivative is 0. 

\begin{equation}
f'(x) = 0
\end{equation}

Take for example again the same function we have been using

\begin{equation}
f(x) = x^3 - 21 x^2 + 135x - 220
\end{equation}

```python
def f(x):
    return x**3 - 21 * x**2 + 135 * x - 220

x = np.linspace(2, 12, 1000)
#x = np.linspace(2.45, 2.47)
plt.plot(x, f(x))
plt.grid();
```

If we take the derivative of the function and set it to 0 to find
the locations of the minima and maxima we have

\begin{equation}
f'(x) = 3x^2 - 42x + 135 \\
3x^2 - 42x + 135 = 0 \\
3(x - 5) (x - 9) = 0
\end{equation}

We could have used the quadratic equation to determine the solution to setting our
derivative to 0, but in this case we can also relatively easily factor the quadratic
expression to easily see where the roots must be located.
Thus wee see that when $x = 5$ or $x = 9$ these will be solutions to this equation,
and from the graph we see that these are indeed the locations of the minimum and the
maximum of the function.

# Integrals of Functions


A definite integral of a continuous function $f(x)$ is a measure of the area under the graph of $f$ in
some interval $[a, b]$.  The integral of a function is related to the derivative of the function.  It is not
too difficult to derive this association.  Lets say you have a continuous function $f(x)$ and you want to
find the area under the curve of this graph in the interval $[a, x]$ from some *fixed* lower value $a$ to
any *variable* upper value $x$.



```python
def f(x):
    """An example of a specific function we want to find the area under the
    graph of.
    """
    return x * (x - 3) * (x - 6) + 20

plt.figure(figsize=(16, 12))
x = np.linspace(0, 6.25)
plt.plot(x, f(x), 'k-', linewidth=3)
plt.grid();

# a = 0, x = 4, deltax = 2
# represent F(x), function to calculate area from fixed a to x
a = 0
b = 4
deltax = 2
z = 5.35
x = np.linspace(a, b)
plt.fill_between(x, f(x), color='lightblue')

# represent F(x + deltax) - F(x)
deltax = 2
x = np.linspace(b, b+deltax)
plt.fill_between(x, f(x), color='lightgreen');

# label the graph
# first remove the specific tics and set general/example tic labels
plt.yticks([0])
plt.xticks([a, b, z, b+deltax], ['$a$', '$x$', '$z$', '$x + \Delta x$'], fontsize=20);

# label the functions
plt.xlim([0, 7.6])
plt.text(2.75, 25, '$y = f(x)$', fontsize=20)
plt.text(1.5, 12, '$F(x)$', fontsize=20)
plt.text(6.25, 15, '$f(z)$', fontsize=20)
plt.arrow(6.25, 15, z-6.05, f(z)-14.4, head_width=0.1, head_length=0.5, color='black')
plt.text(6.25, 7, '$F(x + \Delta x) - F(x)$', fontsize=20)
plt.arrow(6.25, 7, -0.5, 0, head_width=0.2, head_length=0.1, color='black')

# add lines for width and height of area
plt.plot([z, z], [0, f(z)], 'k--')
plt.plot([b, b+deltax], [f(z), f(z)], 'k--');
```

Let $F(x)$ = Area under the graph of $f$ in the interval $[a, x]$.

then $F(x + \Delta x)$ = Area under the graph of $f$ in the interval $[a, x + \Delta x]$

if we take the difference of these two areas we have 

\begin{equation}
F(x + \Delta x) - F(x) = \text{Area under the graph of} f \; \text{in the interval} \; [x, x + \Delta x]
\end{equation}

The area of this difference (the green bit of the figure) has to be equal to the area of a rectangle with height $f(z)$,
where $z$ is some value in the interval $[x, x + \Delta x]$, and with width of $\Delta x$,
thus the area of the previous difference can be written as:

\begin{equation}
F(x + \Delta x) - F(x) = f(z) \Delta x
\end{equation}

rewriting we have

\begin{equation}
\frac{F(x + \Delta x) - F(x)}{\Delta x} = f(z)
\end{equation}

This should look a little familiar.  This is the finite difference equation we used (but with the function $F$) in
developing the concept of the derivative.  If we take this expression in the limit again as $\Delta x \to 0$
we see that the right hand side $f(z)$ becomes $f(x)$ (because z is some value in the interval from
$[x, x + \Delta x]$, and as $\Delta x \to 0$, $z$ simply becomes $x$).  Thus we have

\begin{equation}
\frac{d}{dx} F(x) = f(x)
\end{equation}

$F$ is called the **antiderivative** of $f$.  In other words, $F$ is the function such
that, if you take the derivative of it $F'(x)$ you will end up the the function
$f(x)$.  In essence, if you do the inverse of the process of determining
the derivative of a function, we can use this antiderivative to calculate values of the function $F$ in order to
exactly measure the area under the curve of functions $f$.

We formally express this concept of calculating the area under the graph of a function $f$ from its
antiderivative using the symbol for the definite integral of a function like this:

\begin{equation}
F(x) = \int_a^b f(x) \; dx
\end{equation}

## Properties of Integrals

Integrals possess the analogues of the properties of derivatives.  For example the antiderivative
of a constant and of the sum and difference of functions work in the same way as for derivatives.
The Power Rule, Logarithm Rule and Exponential Rule work basically in an inverted manner to calculate
the antiderivative of a function.

## Calculating Areas using Definite Integrals

If you can calculate the antiderivative of a function $f$, you can use this to calculate the area under
the graph of the curve of the function $f$ on any finite interval $[a, b]$.

For example, take the function

\begin{equation}
f(x) = x^3 (1 - x^4)^2
\end{equation}

The graph of this function in the interval $[0, 1]$ looks like this.

```python
def f(x):
    return x**3 * (1 - x**4)**2

a = 0
b = 1.0
x = np.linspace(a, b, 1000)
plt.plot(x, f(x), 'k-', linewidth=3)
plt.fill_between(x, f(x))
plt.grid();
```

As a rough estimate, since the maximum height of the function is 0.2, we know that the area of the box is 0.2.
The area under the graph looks like it is probably less than $1/2$ of the total area of this bounding box, so
I would expect an area of a bit under 0.1 for this function.

We can calculate the antiderivative of the function.  This function is a polynomial, so it is probably easiest
to simply exapand the terms and then calculate the antiderivative of the sum/difference of the terms, like this:

\begin{equation}
\begin{split}
\int_0^1 x^3 (1 - x^4)^2 \; dx \\
\int_0^1 x^3 (1 - 2x^4 + x^8) \; dx \\
\int_0^1 x^3 - 2x^7 + x^{11} \; dx \\
\end{split}
\end{equation}

The antiderivative of simple terms raised to a power will use the reverse of the Power Rule, ths we have the
antiderivative

\begin{equation}
F(X) = \Big[ \frac{x^4}{4} - \frac{2 x^8}{8} + \frac{x^{12}}{12} \Big]_0^1
\end{equation}

(We leave it as an exercise for the student to prove to yourself that if you
take the derivative $\frac{d}{dx} F(X)$ you get the original function $f(x)$).)

Notice the notation of the 0 and 1 on the expression for the antiderivative here.  To use this
antiderivative to calculate the area under the curve exactly, you take

\begin{equation}
F(b) - F(a)
\end{equation}

thus we can calculate the area of this particular function in the interval $[a=0, b=1]$ as

\begin{equation}
\Big[ \frac{1^4}{4} - \frac{2 \cdot 1^8}{8} + \frac{1^{12}}{12} \Big] - 
\Big[ \frac{0^4}{4} - \frac{2 \cdot 0^8}{8} + \frac{0^{12}}{12} \Big] =
\frac{1}{12} - 0 = \frac{1}{12} = 0.083333
\end{equation}

This is the general procedure for calculating a definite integral symbolically
(an exact solution).  You first need to derive the expression
for the antiderivative, using the inverse of the derivative rules, if you can.  If such an antiderivate
expression is possible, then the area under the curve for any definite interval can be calculated by taking
the difference of the antiderivative on the two ends of the interval.
